```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
frutas = pd.read_csv("fruit_data_with_colors.csv")
frutas.head()
```

```python
#Seleccion de caracteristicas o variables predictoras o variables independientes
x = frutas[["height", "width", "mass", "color_score"]]
#Variable objetivo o variable a predicir
y = frutas["fruit_name"]
```

```python
#Dividir dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
clasificador_KNN = KNeighborsClassifier(n_neighbors=1)
clasificador_KNN.fit(X_train, y_train)
print("Precision del clasificador en fase de entrenamiento", clasificador_KNN.score(X_train, y_train))
```

```python
y_pred = clasificador_KNN.predict(X_test)
print("Reporte de metricas del clasficador: \n,", classification_report(y_test, y_pred))
```

```python
print ("Matriz de confusion: \n", confusion_matrix(y_test, y_pred), "\n")
```

```python
from sklearn.inspection import permutation_importance
result = permutation_importance(clasificador_KNN, X_test, y_test, n_repeats=30, random_state=42)
importancia = pd.DataFrame ({
    "Variable": x.columns,
    "Importancia Media": result.importances_mean,
    "Desviacion": result.importances_std
}).sort_values("Importancia Media", ascending=False)
print(importancia)
```

```python
import plotly.express as px
fig4 = px.bar(importancia, x="Variable", y="Importancia Media",
              error_y="Desviacion", title="Importancia de caracteristicas (KNN - Permutation Importance)",
              text_auto=".3F", color="Variable")
fig4.update_layout(width=600, height=400)
fig4.show()
```

```python
from sklearn.preprocessing import StandardScaler
x = frutas[["height", "width", "mass", "color_score"]]
y = frutas["fruit_name"]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, stratify=y)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(x_train_scaled, y_train)
y_pred_scaledd = knn_scaled.predict(x_test_scaled)
print("\nReporte de clasificacion")
print(classification_report(y_test, y_pred_scaledd))
```

```python
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_scaledd), "\n")
```

```python
from sklearn.inspection import permutation_importance
result = permutation_importance(knn_scaled, x_test_scaled, y_test, n_repeats=30, random_state=42)
importancia = pd.DataFrame ({
    "Variable": x.columns,
    "Importancia Media": result.importances_mean,
    "Desviacion": result.importances_std
}).sort_values("Importancia Media", ascending=False)
print(importancia)
```

```python
import plotly.express as px
fig4 = px.bar(importancia, x="Variable", y="Importancia Media",
              error_y="Desviacion", title="importancia de caracteristicas (KNN - Permutation Importancia)",
              text_auto=".3f", color="Variable")
fig4.update_layout(width=600, height=400)
fig4.show()
```
