```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
import numpy as np
autos = pd.read_csv('AutosDatosSucios.csv')
autos.info()
```

```python
autos.head(10)
```

```python
autos.replace("?", np.nan, inplace= True)
autos.head()
```

```python
autos.info()
```

```python
valores_perdidos = autos.isnull()
valores_perdidos.head(10)
```

```python
# contar valores perdidos en la columna normalized-losses
autos["normalized-losses"].isnull().value_counts()
```

```python
# contar valores perdidos en todas las columnas
autos.isna().sum().sort_values()
```

```python
avg_norm_loss = autos["normalized-losses"].astype("float").mean(axis=0)
print("Promedio de normalized-losses:", avg_norm_loss)
```

```python
autos["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore = autos['bore'].astype('float').mean(axis=0)
print("promedio de bore:", avg_bore)
```

```python
autos["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = autos['stroke'].astype('float').mean(axis=0)
print("promedio de stroke:", avg_stroke)
```

```python
autos["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_horsepower = autos['horsepower'].astype('float').mean(axis=0)
print("promedio de hosepower:", avg_horsepower)
```

```python
autos["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrmp = autos['peak-rpm'].astype('float').mean(axis=0)
print("promedio de peak rpm:", avg_peakrmp)
```

```python
autos['peak-rpm'].replace(np.nan, avg_peakrmp, inplace=True)
```

```python
autos['num-of-doors'].value_counts()
```

```python
#reemplazamos el numero de puertas por el valor mas frecuente
autos["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#Eliminar toda la fila con los valores NaN en la columna "price"
autos.dropna(subset=["price"], axis=0, inplace=True)
#Restablecer el inidce debido a que eliminamos dos filas
autos.reset_index(drop=True, inplace=True)
#verificamos
autos.head(10)
```

```python
autos[["bore", "stroke"]] = autos[["bore", "stroke"]].astype("float")
autos[["normalized-losses"]] = autos[["normalized-losses"]].astype("int")
autos[["price"]] = autos[["price"]].astype("float")
autos[["peak-rpm"]] = autos[["peak-rpm"]].astype("float")
```

```python
autos.dtypes
```

```python
#creamos una nueva columna con la conversion de las unidades
#la formula de conversion es
#L/100km = 235 / m.p.g
autos['city-L/100km']= 235 / autos["city-mpg"]
```

```python
#revisar los datos tranformados
#ver al finar que se agrega la nueva columna
autos.head()
```

```python
#Visualizamos las columnas a normalizar
autos[["length", "width", "height"]].head()
```

```python
#intercambiamos el valor original por el (valor original) / (valor maximo)
autos['length'] = autos['length'] / autos['length'].max()
autos['width'] = autos['width'] / autos['width'].max()
autos['height'] = autos['height'] / autos['height'].max()
```

```python
#Volvemos a verificar
autos[["length", "width", "height"]].head()
```

```python
# Usando cuando las categorias no tiene significado de importancia o niveles
# primero verificamos los tipos de combustible que existen
autos['fuel-type'].value_counts()
```

```python
#Codificamos las variables categoricas usando la libreria sklearn usando el metodo LabelEncoder
#labelEnconder = codifica las variables categoricas de destino con valores entre 0 y n_classes
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#agregamos una nueva columna donde codificaremos el tipo de gasolina a variabke numerica
autos['fuel-type_code'] = le.fit_transform(autos['fuel-type'])
autos.head()
```

```python
#verificamos la culumna que acabamos de agregar
autos['fuel-type_code'].value_counts()
```

```python
variable_dummy1 = pd.get_dummies(autos["fuel-type"], dtype=int)
variable_dummy1.head()
```

```python
#combinar el dataframe "autos" y el el dataframe "variable_dummy1"
autos = pd.concat([autos, variable_dummy1], axis=1)
autos.head(10)
```

```python
mapeo_dicc = {'diesel' :0, 'gas':1}
autos['fuel-type'].map(mapeo_dicc)
autos.head()
```

```python
autos["horsepower"] = autos["horsepower"].astype(int, copy=True)

plt.hist(autos["horsepower"])

# establecer las etiquetas x/y y muestra el titulo
plt.xlabel("caballo de fuerza")
plt.ylabel("cantidad")
plt.title("Histograma de horsepower")

```

```python
#establecebemos el valor de bins (contenedores) igual a 3
plt.hist(autos["horsepower"], bins = 3 )
#establece las etiquetas x/y y muestra el titulo
plt.xlabel("caballos de fuerza")
plt.ylabel("cantidad")
plt.title("contedores horsepower")
```

```python
#contenedores del rango de valores
bins = np.linspace( min( autos["horsepower"]), max( autos["horsepower"] ), 4)
bins
```

```python
#definamos los nombres de grupo que seran las variables categoricas de acuerdo al valor
#de horsepower
grupos = ['low', 'medium', 'higth']
```

```python
#Definamos una nueva columna con las categorias
autos['horsepower-binned'] =pd.cut(autos['horsepower'], bins, labels=grupos, include_lowest=True)
#verficamos el etiquetado de valores
autos[['horsepower', 'horsepower-binned']].head(10)
```

```python
#verificamos las caracteristicas de la nueva columna de categorias
autos['horsepower-binned'].value_counts()
```

```python
#volvemos a graficas la distibucion de la nueva columna etiquetada de categorias
pyplot.bar(grupos, autos["horsepower-binned"].value_counts())
plt.xlabel("Caballos de fuerza")
plt.ylabel("cantidad")
plt.title("contenedores horsepower")
```
